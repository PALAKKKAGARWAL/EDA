##### Robust Scalar
It is used to scale the feature to median and quantiles.
Scaling using median and quantiles consists of substracting the median to all the observations, and then dividing by the interquantile difference. The interquantile difference is the diff between the 75th and 25th quantile.

IOR = 75th quantile - 25th quantile
X_scaled=(X-X.median)/IQR

```python
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
df_robust_scalar=pd.DataFrame(scaler.fit_transform(df),columns=df.columns)
df_robust_scalar.head()
```

```python
plt.hist(df_robust_scalar['Age'],bins=20)
```

##### Guassian Transformation
If your distribution is not normally distributed use can use mathematical operation to convert it into guassion distribution or normal distribution and for this we require guassian transformation.

Some machine learning algo like linear and logistic assume that the features are normally distributed-Accuracy-Performance.
1. Logarithmic transformation
2. Reciprocal transformation
3. suare root transformation
4. exponential transformation
5. boxcox transformation

```python
df=pd.read_csv(r"C:\Users\Asus\Downloads\train.csv", usecols=['Age','Fare','Survived'])
df.head()
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
```

```python
import scipy.stats as stat
import pylab
```

```python
#### If you want to check wheater feature is guassian or normal distribution we use
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'Age')
```

```python
#### Logarithmic Transformation
df['Age_log']=np.log(df['Age'])
plot_data(df,'Age_log')
```

```python
plot_data(df,'Fare')
```

```python
df['Fare_log']=np.log(df['Fare'])
plot_data(df,'Fare_log')
```

```python
#### Reciprocal Transformation
df['Age_resi']=1/df.Age
plot_data(df,'Age_resi')
```

```python
#### Square Root transformation
df['Age_sq']=df.Age**(1/2)
plot_data(df,'Age_sq')
```

```python
#### Exponential Transformation
df['Age_expo']=df.Age**(1/1.2)
plot_data(df,'Age_expo')
```

##### Boxcox Transformation
The Box-Cox transformation is defined as:

T(Y)=(Y exp(λ)−1)/λ

where Y is the response variable and λ is the transformation parameter. λ varies from -5 to 5. In the transformation, all values of λ are considered and the optimal value for a given variable is selected.

```python

df['Age_boxcox'],parameters=stat.boxcox(df['Age'])
```

```python
print(parameters)
```

```python
plot_data(df,'Age_boxcox')
```

```python
#### Fare
df['Fare_log']=np.log1p(df['Fare'])
plot_data(df,'Fare_log')
```

```python
df['Fare_boxcox'],parameters=stat.boxcox(df['Fare']+1)
plot_data(df,'Fare_boxcox')
```

```python
df['Fare_log']=np.log(df['Fare']+1)
plot_data(df,'Fare_log')
```
